# Resultados

A variável com maior impacto é o número de horas de estudo: quanto mais horas estudadas, maior a nota na prova, com correlação de 0,83. O maior impacto negativo está nas horas gastas em redes sociais e na Netflix, ambas com correlação de -0,17. A idade foi a variável menos relevante, com correlação de -0,01. A saúde mental também apresentou relação positiva com a nota, indicando que melhores índices estão associados a melhor desempenho. A frequência aparece como o terceiro fator mais importante (0,16), seguida pelas horas de sono (0,12).


```python
filtro_horas_altas = df["study_hours_per_day"] >= 5

filtro_horas_baixas = df["study_hours_per_day"] <= 4
```

```python
media_alta = df[filtro_horas_altas]["exam_score"].mean()

media_baixa = df[filtro_horas_baixas]["exam_score"].mean()

print(media_alta)
print(media_baixa)
```

```python
print(f"A media das pessoas que estudaram 5 ou mais horas tiveram a media de {media_alta:.2f},enquanto a media de quem estudou 2 ou menos horas tiveram a media de {media_baixa:.2f},Mostrando de quem estudou mais horas tera um nota maior")
```
