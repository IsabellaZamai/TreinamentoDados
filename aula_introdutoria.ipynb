```python
# Verificar a qualidade dsas notas de um aluno
# maior do que 9: exelente
# maior do que 7; boa
# maior do que 5: regular
# Abaixo de 5 : reprovado
```

```python
nota = int(input("Digite a sua nota: "))
print(nota)
```

```python
if nota > 9:
    print("Nota exelente")
elif nota > 7:
    print("Nota boa")
elif nota > 5:
    print("nota regular")
else:
    print("Reprovado")
```
