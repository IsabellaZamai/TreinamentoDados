Criar duas listas:
-5 produtos (apenas inserir valores)
-A primeira lista, custo do produto
-A segunda lista, o preco  de venda
-desafio; Calcular o lucro de cada produto.

```python
CustoP = [5, 4 , 10 , 6 , 3 ]
precoV = [15, 10, 20 ,21 , 7 ]

CustoP_array = np.array(CustoP)
precoV_array = np.array(precoV)

Lucro = precoV_array -CustoP_array
print(f"O lucro sera de: {Lucro}")
```

```python
custo = []
preco = []

while len(custo) & len(preco) < 5:
    valorC = int(input("Diga o custo dos itens: "))
    custo.append(valorC)
    valorP =  int(input("Diga p preco de venda: "))
    preco.append(valorP)


custo_array = np.array(custo)
preco_array = np.array(preco)
lucro = preco_array - custo_array

print("Os seu lucro será", lucro)
```

Criar uma lista de vendas por dia 
vendas = [10,20,45,5,90]
-Exibir a Maior venda 
-exibir a menor venda
- exibir a média da vendas
-exibir a mediana de vendas

```python
vendas = np.array([10,20,45,5,90])

maior = np.max(vendas)
menor = np.min(vendas)
media = np.mean(vendas)
mediana = np.median(vendas)

print("="*64)
print("a lista é:")
print(vendas)
print("="*64)
print(f"o maior é {maior}, o menor é {menor} a media é {media} e a mediana é {mediana}")
print("="*64)
```
